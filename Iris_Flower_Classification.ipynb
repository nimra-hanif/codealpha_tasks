**Iris Flowers Measurements**

```python
import pandas as pd

# Features (measurements)
feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = df[feature_names]

# Target (flower type)
y = df['Species']

# Create a new DataFrame with features and target for better understanding
df_processed = pd.DataFrame(X, columns=feature_names)
df_processed['species'] = y

print(df_processed.head())
```

```python
print(df_processed[df_processed['species'] == 'Iris-versicolor'].head())
print(df_processed[df_processed['species'] == 'Iris-virginica'].head())
```

**Machine Learning Model**

```python
# Import
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load data
iris = load_iris()
X = iris.data      # measurements
y = iris.target    # species

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = LogisticRegression()
model.fit(X_train, y_train)

# Test prediction
predictions = model.predict(X_test)

print(predictions)
```

**Accuracy**

```python
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, predictions))
```

**Performance**

```python
from sklearn.metrics import classification_report, confusion_matrix

# Detailed performance
print("\nClassification Report:\n", classification_report(y_test, predictions))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, predictions))
```

**Basic Classification Concept** in Machine Learning

Classification is when a model learns to assign a label (category) to data.

Input → flower measurements (length & width)

Output → species:
Setosa, Versicolor and Virginica

**Types of Classification**

There are 2 types of classification **Binary** and **Multi-class** Classification. Binary means only 2 classes and Multi-class means more than 2 classes such as Iris dataset (3 flowers).

**Features** is known as (X) the input data such as sepal length, petal width while **Target/Label** is known as (y) which refers to the output.

**Training Data** is used to teach the model while **Testing Data** is used to check performance.

**Performance** is checked through **Accuracy** which shows the percentage of the correct prediction.

**Confusion Matrix** shows the correct and wrong predictions.
